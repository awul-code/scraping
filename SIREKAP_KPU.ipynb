```python
from google_play_scraper import Sort, app, reviews_all
import pandas as pd
import numpy as np

result = reviews_all(
    "id.go.kpu.sirekap2024",
    sleep_milliseconds=0,
    lang="id",
    country="id",
    sort=Sort.NEWEST,
)

df_busu = pd.DataFrame(np.array(result), columns=["review"])
df_busu = df_busu.join(pd.DataFrame(df_busu.pop("review").tolist()))
df_busu.head()
```

```python
# Memunculkan komentar yang baru
new_df = df_busu[['userName','score','at','content']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score', 'at', 'content']]
my_df = my_df[['content','score']]
my_df.head()
```

```python
my_df = df_busu[['score', 'content']]
def pelabelan(score):
    if score < 3:
        return "Negatif"
    elif score == 4:
        return "Positif"
    elif score == 5:
        return "Sangat Positif"

my_df['Label'] = my_df['score'].apply(pelabelan)
my_df.head(10)
```

```python
my_df.to_csv('scrapped_data.csv', index=False)
```
